```python
import pandas as pd
cereal = pd.read_csv('cereal.csv')
cereal.columns = ['name','mfr','type','calories','protein',
                  'fat','sodium','fiber','carbo','sugars',
                  'potass','vitamins','shelf','weight','cups','rating']
#刪除不必要資料
DropString = cereal.drop('name',1)
DropString = DropString.drop('mfr',1)
DropString = DropString.drop('type',1)

'''
#直接取特定的值
DropString = cereal[['calories','protein','fat','sodium','fiber','carbo','sugars',
                  'potass','vitamins','shelf','weight','cups','rating']]
'''
#處理特殊值
for i in range(0, len(DropString)):
    if DropString['potass'][i] == -1:
        DropString['potass'][i] = 0

DropString
```

```python
#訓練賦值 測試賦值
x = DropString[['calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars',
       'potass', 'vitamins', 'shelf', 'weight', 'cups']]
y = DropString[['rating']]
```

```python
#以 rating 取平均作為預測分類依據
for i in range(0, len(DropString)):
    if float(y['rating'][i]) > float(y.mean()):
        y['rating'][i] = 1
    else:
        y['rating'][i] = 0
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(criterion = 'entropy', max_depth=3, random_state=0)
tree.fit(X_train,y_train)
```

```python
error = 0
for i, v in enumerate(tree.predict(X_test)):
    if v!= y_test.values[i]:
        print(i,v)
        error+=1
print(error)
```

```python
tree.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree, x, y, cv=10) #cross validation
scores

print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
from sklearn.tree import export_graphviz
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from sklearn import metrics

export_graphviz(tree, out_file='tree.dot', feature_names = x.columns, class_names = ['Floor', 'Top'])

!dot -Tpng tree.dot -o tree.png

%matplotlib inline
img = mpimg.imread('tree.png')
fig = plt.figure(figsize=(18,20))
plt.imshow(img)
```

```python
print(metrics.classification_report(y_test, tree.predict(X_test)))
print(metrics.confusion_matrix(y_test, tree.predict(X_test),labels=[0, 1]))

tree.feature_importances_.tolist()
df = pd.DataFrame({'feature':x.columns,'feature_importance':tree.feature_importances_.tolist()})
df = df.sort_values(by=['feature_importance'],ascending=False).reset_index(drop=True)
df
```

```python
#隨機森林
from sklearn.ensemble import RandomForestClassifier
 
rf = RandomForestClassifier(criterion='gini', 
                             n_estimators=1000,
                             min_samples_split=12,
                             min_samples_leaf=1,
                             oob_score=True,
                             random_state=1,
                             n_jobs=-1) 
#(你的訓練變數 X).iloc[:, 0:], (你的測試變數 Y).iloc[:, 0]
rf.fit(x.iloc[:, 0:], y.iloc[:, 0])
print("%.4f" % rf.oob_score_)
```

```python
#(你的訓練變數 X).iloc[:, 0:].columns, columns = ['variable'] variable只是上面的欄位名稱
pd.concat((pd.DataFrame(x.iloc[:, 0:].columns, columns = ['feature']), 
           pd.DataFrame(rf.feature_importances_, columns = ['feature_importance'])), 
          axis = 1).sort_values(by='feature_importance', ascending = False)[0:]
#[0:] 可寫可不寫，都是顯示全部 ; [:5]表示顯示5個
```
