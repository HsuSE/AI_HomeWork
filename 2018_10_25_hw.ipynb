```python
import warnings
warnings.filterwarnings("ignore")
#讀取檔案資料
import pandas as pd
cereal = pd.read_csv('cereal.csv')

# GetData = cereal[['calories','protein','fat','sodium','fiber','carbo','sugars',
#                   'potass','vitamins','shelf','weight','cups','rating']]

#處理特殊值
for i in range(0, len(cereal)):
    if cereal['potass'][i] == -1:
        cereal['potass'][i] = 0
        
#保有原始迴歸        
GetData_regression = cereal[['calories','protein','fat','sodium','fiber','carbo','sugars','potass','vitamins','shelf','weight','cups','rating']]        

#以 rating 取平均作為預測分類依據 (類別)
for i in range(0, len(cereal)):
    if float(cereal['rating'][i]) > float(cereal['rating'].mean()):
        cereal['rating'][i] = 1
    else:
        cereal['rating'][i] = 0
```

```python
#訓練賦值 測試賦值 (類別用)
Train_x = cereal[['calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars','potass', 'vitamins', 'shelf', 'weight', 'cups']]
Text_y = cereal[['rating']]
#訓練賦值 測試賦值 (回歸用)
Train_x_regression = GetData_regression[['calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars','potass', 'vitamins', 'shelf', 'weight', 'cups']]
Test_y_regression = GetData_regression[['rating']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
X_train, X_test, y_train, y_test=train_test_split(Train_x_regression, Test_y_regression, test_size=0.3, random_state=0)

#資料標準化
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.fit_transform(X_test)
y_train_std = sc.fit_transform(y_train)
y_test_std = sc.fit_transform(y_test)

from sklearn import svm
X = X_train_std 
y = y_train_std 
kernel_type=['','linear','poly','rbf','sigmoid']
for i in range(1,5):
    clf = svm.SVR(C=i,gamma=(i/100),kernel=kernel_type[i])
    print("參數\n%s" % clf.fit(X, y))
    print("訓練準確度\n%s" % clf.score(X,y))
    print("測試準確度\n%s" % clf.score(X_test_std,y_test_std))
    print("交叉測試準確度\n%s" %cross_val_score(clf,X,y, cv=Cross_Time).mean())
    print("----------------------")
```
