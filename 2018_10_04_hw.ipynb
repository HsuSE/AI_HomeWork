```python
import pandas as pd
cereal = pd.read_csv('cereal.csv')
cereal.columns = ['name','mfr','type','calories','protein',
                  'fat','sodium','fiber','carbo','sugars',
                  'potass','vitamins','shelf','weight','cups','rating']
#刪除不必要資料
DropString = cereal.drop('name',1)
DropString = DropString.drop('mfr',1)
DropString = DropString.drop('type',1)

'''
#直接取特定的值
DropString = cereal[['calories','protein','fat','sodium','fiber','carbo','sugars',
                  'potass','vitamins','shelf','weight','cups','rating']]
'''
#處理特殊值
for i in range(0, len(DropString)):
    if DropString['potass'][i] == -1:
        DropString['potass'][i] = 0

DropString
```

```python
#訓練賦值 測試賦值
x = DropString[['calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars',
       'potass', 'vitamins', 'shelf', 'weight', 'cups']]
y = DropString[['rating']]
```

```python
#以 rating 取平均作為預測分類依據
for i in range(0, len(DropString)):
    if float(y['rating'][i]) > float(y.mean()):
        y['rating'][i] = 1
    else:
        y['rating'][i] = 0
y
```

```python
def callrandomforest(treecount,samples_split):
    #隨機森林
    from sklearn.ensemble import RandomForestClassifier
    #n_estimators為樹木的數量(投票數量)
    rf = RandomForestClassifier( criterion = 'gini', 
                                 n_estimators = treecount,
                                 min_samples_split = samples_split,
                                 min_samples_leaf = 1,
                                 oob_score = True,
                                 random_state = 1,
                                 n_jobs = -1) 
    #(你的訓練變數 X).iloc[:, 0:], (你的測試變數 Y).iloc[:, 0]
    rf.fit(x.iloc[:, 0:], y.iloc[:, 0])
    print(">>>>>>預測準確度為：%.4f" % rf.oob_score_)
    
    #(你的訓練變數 X).iloc[:, 0:].columns, columns = ['variable'] variable只是上面的欄位名稱
    pd.concat((pd.DataFrame(x.iloc[:, 0:].columns, columns = ['feature']),  pd.DataFrame(rf.feature_importances_, 
    columns = ['feature_importance'])), axis = 1).sort_values(by='feature_importance', ascending = False)[0:]
    #[0:] 可寫可不寫，都是顯示全部 ; [:5]表示顯示5個
    
    kk = pd.concat((pd.DataFrame(x.iloc[:, 0:].columns, columns = ['feature']),  pd.DataFrame(rf.feature_importances_, 
    columns = ['F_importance'])), axis = 1).sort_values(by='F_importance', ascending = False)[:5]
    print(kk)

    
```

```python
for votenumber in(20,40,60):
    for samples_split in (5,10,15):
        print('當前決策數樹目：',votenumber,'; 當前min_samples_split：',samples_split)
        callrandomforest(votenumber,samples_split)
```
