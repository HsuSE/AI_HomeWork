```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
class linear_model(object):
    
    def __init__(self, lr=0.006):
        self.lr = lr
        self.create_sample()
        self.w_init()
    
    def w_init(self):
        self.w = np.random.uniform(-10, 10, 1)
        self.bias = np.random.uniform(-10, 10, 1)
        return 0
    
    def create_sample(self, num=100):
        self.Train_X = np.random.randint(100, 210, 100)
        self.Train_y = np.add(np.random.uniform(-20, 20, 100), np.multiply(self.Train_X, 2))

    def visualized(self):
        plt.plot(self.Train_X, self.Train_y, 'b.')
        l = [x for x in range(100, 210)]
        plt.plot(l, self.predict(l), '-r')
        plt.show()
        
    def loss(self, Y, y):
        return np.sum(np.power(Y-y, 2))/2.
    
    def GD_optimizer(self, epoch):
        lr = self.lr - epoch*0.000004
        x0 = np.mean((self.predict(self.Train_X) - self.Train_y) * self.Train_X)/np.mean(self.Train_X)
        x1 = np.mean(self.predict(self.Train_X) - self.Train_y)
        self.w -= self.lr * x0
        self.bias -= self.lr * x1
    
    def fit(self, epochs=50):
        pre_loss = self.loss(self.Train_y, self.predict(self.Train_X))
        print("loss: %.2f, Weight: %2.2f, Bias: %2.2f" % (pre_loss, self.w, self.bias))
        min_loss = 99999
        count = 0
        step = 0
        for epoch in range(epochs):
            pre_loss = self.loss(self.Train_y, self.predict(self.Train_X))
            self.GD_optimizer(epoch)
            loss = self.loss(self.Train_y, self.predict(self.Train_X))
            print("Epoch: %3d \nloss: %.2f, Weight: %2.2f, Bias: %2.2f" % (epoch+1, loss, self.w, self.bias))
            if min_loss > loss:
                min_loss = loss
                count += 0
            else:
                count += 1
            if count > 500:
                break
            if loss < 1.0:
                break
            step = epoch
        return "Epoch: %3d \nloss: %.2f, Weight: %2.2f, Bias: %2.2f" % (epoch+1, loss, self.w, self.bias)
    
    def predict(self, X):
        Y = np.add(np.multiply(X, self.w), self.bias)
        return Y
```

```python
model = linear_model()
```

```python
model.visualized()
```

```python
result = model.fit(1000000)
```

```python
model.visualized()
print(result)
```
